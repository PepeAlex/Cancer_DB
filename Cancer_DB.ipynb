<h1 style='text-align: center'> Cancer Deaths by US-State</h1>

I found this data in the site below:

https://corgis-edu.github.io/corgis/csv/cancer/

The data has 51 rows that correspond to US-States and 75 columns divided by numbers of Cancer Deaths subdivided by Race, Years, Sex and three kinds of cancer: Breast, Colorectal and Lung. 

I made some graphs to describe the data.

Some libraries were imported as:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the csv file
df = pd.read_csv("cancer.csv")
# print the first 5 rows
df.head()
```

```python
# Graph using Matplotlib 
plt.rcParams['figure.figsize'] = (20, 7)
ax = sns.barplot(x = df["State"], y = df["Total.Population"], palette= 'dark')
ax.set_xlabel(xlabel = "State", fontsize = 10)
ax.set_ylabel(ylabel = "Total.Population" , fontsize = 10)
ax.set_title(label = "Population deaths of each US-State", fontsize = 20)
plt.xticks(rotation = 45)
plt.show()
```

```python
# add some attributes and reasign with df_age
df_age = df['Rates.Age.18-45'] + df["Rates.Age.45-64"] + df["Rates.Age.> 64"]
```

```python
# more graph
plt.rcParams['figure.figsize'] = (12, 20)
ax = sns.barplot(x = df_age , y = df["State"], palette= 'dark')
ax.set_xlabel(xlabel = "State", fontsize = 10)
ax.set_ylabel(ylabel = "States" , fontsize = 10)
ax.set_title(label = "Numbrers of Adult deaths(age > 18)", fontsize = 20)
plt.xticks(rotation = 90)
plt.show()
```

```python
print("Population deaths by Ages Rate")

sns.pairplot(df, x_vars=['Total.Population', 'Total.Population'], y_vars=['Rates.Age.18-45', 'Rates.Age.45-64'])

plt.show()
```

# Race

In this part I will work with human race and illustrate the number of deaths by kind of Cancer.

###Â White Race

```python
white = df['Types.Lung.Race.White'] + df['Types.Colorectal.Race.White'] + df['Types.Lung.Race.White']

total_Breast_White = df['Types.Breast.Race.White'].sum()
total_Colorectal_White = df['Types.Colorectal.Race.White'].sum()
total_Lung_White = df['Types.Lung.Race.White'].sum()
```

```python
labels = ['Breast', 'Colorectal', 'Lung']
sizes = total_Breast_White, total_Colorectal_White, total_Lung_White
explode = (0.0, 0.0, 0.0)

fig1, ax1 = plt.subplots(figsize=(12, 8))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=False, startangle=45)
ax1.axis('equal')
plt.title("Deaths by Cancer in White Race\n")

plt.show()
```
